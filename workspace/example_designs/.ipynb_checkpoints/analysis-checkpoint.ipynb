```python
import os
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
memory_bandwidth = {
    "5060": 448,
    "5080": 960,
    "RTX Pro 6000": 1_790
}

memory_bandwidth = { # GFLOPS
    "5060": 299.6,
    "5080": 879.3,
    "RTX Pro 6000": 1_968
}

boost_ghz = {
    "5060": 2.50,
    "5080": 2.62,
    "RTX Pro 6000": 2.617
}
```

```python
df = pd.read_csv('RTX Pro 6000_ViT_layer_stats.csv')  # Replace with your filename

# Assume GFLOPs are given at 1GHz => GFLOPs/sec directly usable
df['Throughput'] = df['GFLOPs']
df['Total_energy'] = df['fJ/Compute'].apply(lambda x: ast.literal_eval(x)['Total'])

# Extract total compute and assume 2 FLOPs per MAC
df['Compute_ops'] = df['Computes'].astype(float)
df['Bytes'] = df['Energy (uJ)'] * 1e-6 / 1e-9  # Assume 1nJ per Byte (placeholder)
df['Intensity'] = df['Compute_ops'] / df['Bytes']  # X-axis

# Plot setup
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xscale('log')
ax.set_yscale('log')
x_vals = np.logspace(-2, 3, 500)

# Plot rooflines
for accelerator, bw in memory_bandwidth.items():
    y_vals = np.minimum(bw * x_vals, memory_bandwidth[accelerator])
    ax.plot(x_vals, y_vals, label=f'{accelerator} ({bw} GB/s)', linestyle='--')

# Plot measured points
ax.scatter(df['Intensity'], df['Throughput'], color='orange', zorder=5)
for _, row in df.iterrows():
    ax.text(row['Intensity'], row['Throughput'], row['layer'], fontsize=8)

# Labels and aesthetics
ax.set_xlabel('Operational Intensity (FLOPs / Byte)', fontsize=12)
ax.set_ylabel('Throughput (GFLOPs / sec)', fontsize=12)
ax.set_title('Roofline Plot of Neural Network Layers', fontsize=14)
ax.legend()
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
